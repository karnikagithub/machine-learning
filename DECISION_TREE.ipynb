```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
df
```

                    GET DUMMIES IS UISED TO CONVERT CATEGORIAL DATA INTO CONTINUES

```python
df =  pd.get_dummies(df, drop_first = True)   #### it took gender
df
```

```python
x = df.iloc[:,[1,2,4]].values
y = df.iloc[:,-2].values
```

```python
x
```

                    SPLITTING DATA INTO TRAIN_TEST

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.25, random_state = 0)
```

```python
x_train
```

                    STANDARIZATION/FEATURE SCALING OF DATA

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_train
```

                                  IMPORTING OUR MODEL(DECISION TREE)

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
y_pred
```

                CHECKING ACCURACY_SCORE, CONFUSION_MATRIX

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test, y_pred)
```

```python
metrics.mean_squared_error(y_test, y_pred)
```

```python
from matplotlib.colors import ListedColormap
x_set, y_set = sc.inverse_transform(x_train), y_train
x1, x2 = np.meshgrid(np.arange(start = x_set[:,0].min() - 1, stop = x_set[:,0].max() + 1, step =0.01),
                     np.arange(start = x_set[:,1].min() - 1, stop = x_set[:,1].max() + 1, step =0.01))
plt.contourf(x1,x2, classifier.predict(sc.transform(np.array([x1.ravel(), x2.ravel()]).T)).reshape(x1.shape), 
             alpha = 0.75, cmap = ListedColormap('red','green'))
plt.xlim(x1.min(),x1.max())
plt.ylim(x2.min(),x2.max())
for i, j in enumerate(np.unique(y_set)):
  plt.scatter(x_set[y_set == j,0], x_set[y_set == j,1], c = ListedColormap(('red','green'))(i), label =j)
plt.title('DecisionTreeClassifier')
plt.xlabel('Age')
plt.ylabel('Est salary')
plt.legend()
plt.show()
```
